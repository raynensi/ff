```python
# fill missing values with
data_1 = {
    'ID': ['001', '002', '003', '004', '005', '002', np.nan],
    'Name': ['Alice', 'Bob', 'Charlie', None, 'Eve', 'Bob', 'Frank'],
    'Age': ['25', '30', '22', '28', np.nan, '30', '35'],
    'Salary': [50000, np.nan, 45000, 55000, 60000, 52000, 48000],
    'City': ['NYC', 'LA', 'NYC', 'Chicago', 'SF', 'LA', 'NYC']
}
data_1
```

```python
data_1= pd.DataFrame(data_1)
data_1
```

```python
data_1c= data_1.copy()
```

```python
data_1c
```

```python
data_1c['Salary'].fillna(data_1c['Salary'].mean())
```

```python
data_1c['Name'].fillna('unknown')
```

```python
data_1c['ID'].fillna('missing')
```

```python
data_1c.dropna()
```

```python
# identify and remove duplicates
duplicates = data_1c.duplicated()
print(duplicates)
```

```python
data_1c.drop_duplicates()
```
